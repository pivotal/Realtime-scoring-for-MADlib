```python
%%sql

DROP TABLE IF EXISTS madlibflowdemo.raw_transactions;
CREATE TABLE madlibflowdemo.raw_transactions (
    account_id integer,
    account_number text,
    card_type text,
    fraud_flag boolean,
    location_id integer,
    merchant_city text,
    merchant_city_alias text,
    merchant_lat double precision,
    merchant_long double precision,
    merchant_name text,
    merchant_state varchar(2),
    posting_date text,
    rlb_location_key text,
    transaction_amount double precision,
    transaction_date text,
    transaction_id integer
);

COPY madlibflowdemo.raw_transactions FROM '/home/gpadmin/transactions_1546276024362444.csv' CSV HEADER QUOTE '"';

DROP TABLE IF EXISTS madlibflowdemo.transactions;
CREATE TABLE madlibflowdemo.transactions AS
SELECT 
    account_id,
    account_number,
    card_type,
    fraud_flag,
    location_id,
    merchant_city,
    merchant_city_alias,
    merchant_lat,
    merchant_long,
    merchant_name,
    merchant_state,
    to_timestamp(posting_date::float) AT TIME ZONE 'EST' AS posting_date,
    rlb_location_key,
    CASE WHEN transaction_amount < 0 THEN 0 ELSE transaction_amount END AS transaction_amount,
    to_timestamp(transaction_date::float) AT TIME ZONE 'EST' AS transaction_date,
    transaction_id
FROM madlibflowdemo.raw_transactions
DISTRIBUTED RANDOMLY;

SELECT count(*) FROM madlibflowdemo.transactions;
```

```python
%%sql

DROP TABLE IF EXISTS madlibflowdemo.raw_accounts;
CREATE TABLE madlibflowdemo.raw_accounts (
    raw_accounts json
);

COPY madlibflowdemo.raw_accounts FROM '/home/gpadmin/accounts.json';

DROP TABLE IF EXISTS madlibflowdemo.accounts;
CREATE TABLE madlibflowdemo.accounts (
    account_number text,
    expiration_date text,
    cvv text,
    card_type text,
    account_city text,
    account_city_alias text,
    account_state text,
    account_long float,
    account_lat float,
    account_transaction_radius integer,
    trxn_mean float,
    trxn_std float,
    account_id integer
);

INSERT INTO madlibflowdemo.accounts (
    SELECT raw_accounts->>'account_number' AS account_number
          ,raw_accounts->>'expiration_date' AS expiration_date
          ,raw_accounts->>'cvv' AS cvv
          ,raw_accounts->>'card_type' AS card_type
          ,raw_accounts->>'city' AS account_city
          ,raw_accounts->>'city_alias' AS account_city_alias
          ,(raw_accounts->>'state')::varchar(2) AS account_state 
          ,(raw_accounts->>'long')::float AS account_long
          ,(raw_accounts->>'lat')::float AS account_lat
          ,(raw_accounts->>'transaction_radius')::integer AS account_transaction_radius
          ,(raw_accounts->>'trxn_mean')::float AS trxn_mean
          ,(raw_accounts->>'trxn_std')::float AS trxn_std
          ,(raw_accounts->>'account_id')::integer AS account_id
    FROM (
        SELECT json_array_elements(raw_accounts)AS raw_accounts
        FROM madlibflowdemo.raw_accounts
    ) foo
);

SELECT count(*) FROM madlibflowdemo.accounts;
```

```python

def merchant_profile(location_id):
    query = """
        SELECT location_id
              ,merchant_name
              ,merchant_state
              ,merchant_city
              ,merchant_city_alias
              ,merchant_lat
              ,merchant_long
              ,min(extract('hour' from transaction_date)) AS min_hour
              ,max(extract('hour' from transaction_date)) AS max_hour
              ,avg(transaction_amount) AS avg_transaction_amount
              ,min(transaction_amount) AS min_transaction_amount
              ,max(transaction_amount) AS max_transaction_amount
              ,stddev(transaction_amount) AS stddev_transaction_amount
              ,count(*) AS number_transactions
              ,avg(account_transaction_radius) AS avg_account_transaction_radius
              ,count(distinct account_id) AS unique_accounts
              ,count(CASE WHEN fraud_flag = True THEN 1 ELSE null END) AS fraud_cases
              ,count(CASE WHEN card_type = 'Discover' THEN 1 ELSE null END) AS discover_transactions
              ,count(CASE WHEN card_type = 'Amex' THEN 1 ELSE null END) AS amex_transactions
              ,count(CASE WHEN card_type = 'Diners' THEN 1 ELSE null END) AS diners_transactions
              ,count(CASE WHEN card_type = 'MasterCard' THEN 1 ELSE null END) AS mastercard_transactions
              ,count(CASE WHEN card_type = 'Visa' THEN 1 ELSE null END) AS visa_transactions
        FROM madlibflowdemo.transactions_accounts
        WHERE location_id = {}
        GROUP BY 1,2,3,4,5,6,7
    """.format(location_id)
    cur.execute(query)
    
    
    colnames = [desc[0] for desc in cur.description]
    df = pd.DataFrame(cur.fetchall(), columns=colnames)  
    
    message = "<span style='color:black'>**Location ID:** {}</span>".format(df['location_id'].to_string(index=False))
    printmd(message) 

    message = "<span style='color:black'>**Merchant Mame:** {}</span>".format(df['merchant_name'].to_string(index=False))
    printmd(message) 

    message = "<span style='color:black'>**Location:** {}, {}</span>".format(df['merchant_city_alias'].to_string(index=False), df['merchant_state'].to_string(index=False))
    printmd(message) 
    
    return df

merchant_profile(5)

```

<a id="de_continuous"></a>
#### Continuous Columns

```python
contColumns = ['transaction_amount','account_transaction_radius']
sliderValue = 20
colName = contColumns[0]

def get_cont_data_frame(col, buckets):
    query = """
        WITH aggs AS (
            SELECT min({c}) AS min,
                   max({c}) AS max
              FROM madlibflowdemo.transactions_accounts
        )
        SELECT width_bucket({c}, min, max, {b}-1) AS bucket,
               ('[' || min({c}) || ',' || max({c}) || ')')::text as range,
               count(*) as freq
        FROM madlibflowdemo.transactions_accounts, aggs
        GROUP BY bucket
        ORDER BY bucket
    """.format(c=col, b=buckets)
    cur.execute(query)

    colnames = [desc[0] for desc in cur.description]
    return pd.DataFrame(cur.fetchall(), columns=colnames)
    
def graph_reset():
    ipd.clear_output()
    printmd("-----\n")
    ipd.display(widgets.HBox((contDropdown,bucketsSlider)))
    printmd("-----\n")
    df = get_cont_data_frame(colName,sliderValue)
    bar_plot(df,colName,"range",colName,"freq","Frequency", "#4378E2")    
    
def on_cont_selection(res):
    global colName
    if res['type'] == 'change' and res['name'] == 'value':
        colName = res['new']
        graph_reset()
        
def on_slider_selection(res):
    global sliderValue
    if res['new'] == {} and res['old']:
        sliderValue = res['old']['value']
        graph_reset()
    
# Look at log transforms
colsAddLogs = contColumns + ["log({} + 1)".format(c) for c in contColumns]
#colsAddLogs = contColumns

contDropdown = widgets.Dropdown(
    options=colsAddLogs,
    value=colsAddLogs[0],
    description='Column:',
    disabled=False,
)

bucketsSlider = widgets.IntSlider(
    value=sliderValue,
    min=5,
    max=50,
    step=1,
    description='# Buckets:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

contDropdown.observe(on_cont_selection)
bucketsSlider.observe(on_slider_selection)

graph_reset()

```

## Feature Engineering

<a id="fe_transaction"></a>
#### Transaction Features

```python
%%sql

DROP TABLE IF EXISTS madlibflowdemo.transactions_features;
CREATE TABLE madlibflowdemo.transactions_features AS
SELECT transaction_id
      ,transaction_amount
      ,account_id
      ,location_id
      ,fraud_flag
FROM madlibflowdemo.transactions
WHERE transaction_date > now() - interval '30 days';
```

<a id="fe_merchant"></a>
#### Merchant Features

```python
%%sql

DROP TABLE IF EXISTS madlibflowdemo.merchant_features;
CREATE TABLE madlibflowdemo.merchant_features AS
SELECT l.merchant_name
      ,t.*
FROM madlibflowdemo.locations AS l
JOIN (
  SELECT location_id
        ,merchant_state
        ,count(CASE WHEN fraud_flag = True THEN 1 ELSE null END) AS m_fraud_cases
        ,min(extract('hour' from transaction_date)) AS m_min_hour
        ,max(extract('hour' from transaction_date)) AS m_max_hour
        ,avg(transaction_amount) AS m_avg_transaction_amount
        ,min(transaction_amount) AS m_min_transaction_amount
        ,max(transaction_amount) AS m_max_transaction_amount
        ,stddev(transaction_amount) AS m_stddev_transaction_amount
        ,count(*) AS m_number_transactions
        ,count(distinct account_id) AS m_unique_accounts
        ,coalesce(count(CASE WHEN card_type = 'Discover' THEN 1 ELSE null END) / (count(*))::float,0) AS m_prop_discover_transactions
        ,coalesce(count(CASE WHEN card_type = 'Amex' THEN 1 ELSE null END) / (count(*))::float,0) AS m_prop_amex_transactions
        ,coalesce(count(CASE WHEN card_type = 'Diners' THEN 1 ELSE null END) / (count(*))::float,0) AS m_prop_diners_transactions
        ,coalesce(count(CASE WHEN card_type = 'MasterCard' THEN 1 ELSE null END) / (count(*))::float,0) AS m_prop_mastercard_transactions
        ,coalesce(count(CASE WHEN card_type = 'Visa' THEN 1 ELSE null END) / (count(*))::float,0) AS m_prop_visa_transactions
  FROM madlibflowdemo.transactions
  WHERE transaction_date > now() - interval '30 days'
  GROUP BY 1, 2
) AS t
USING (location_id);
```

<a id="fe_account"></a>
#### Account Features

```python
%%sql

DROP TABLE IF EXISTS madlibflowdemo.account_features;
CREATE TABLE madlibflowdemo.account_features AS
SELECT a.account_state
      ,a.card_type
      ,t.*
FROM madlibflowdemo.accounts AS a
JOIN (
  SELECT account_id
        ,min(extract('hour' from transaction_date)) AS a_min_hour
        ,max(extract('hour' from transaction_date)) AS a_max_hour
        ,avg(transaction_amount) AS a_avg_transaction_amount
        ,min(transaction_amount) AS a_min_transaction_amount
        ,max(transaction_amount) AS a_max_transaction_amount
        ,stddev(transaction_amount) AS a_stddev_transaction_amount
        ,count(*) AS a_number_transactions
  FROM madlibflowdemo.transactions
  WHERE transaction_date > now() - interval '30 days'
  GROUP BY 1
) AS t
USING (account_id);
```

<a id="fe_join"></a>
#### Join feature tables

```python
%%sql

DROP TABLE IF EXISTS madlibflowdemo.all_features;
CREATE TABLE madlibflowdemo.all_features AS
SELECT *
      ,abs(t.transaction_amount - m.m_avg_transaction_amount) AS m_transaction_delta
      ,abs(t.transaction_amount - a.a_avg_transaction_amount) AS a_transaction_delta
FROM madlibflowdemo.transactions_features AS t
JOIN madlibflowdemo.merchant_features AS m
USING (location_id)
JOIN madlibflowdemo.account_features AS a
USING (account_id);

SELECT * 
FROM madlibflowdemo.all_features
LIMIT 5;
```

<a id="fe_one_hot"></a>
#### One Hot Encode Transaction Features

https://madlib.apache.org/docs/latest/group__grp__encode__categorical.html

```python
categoricalFeatures = ['merchant_name','merchant_state','m_min_hour','m_max_hour','account_state','card_type','a_min_hour','a_max_hour']
continuosFeatues = ['transaction_amount','m_fraud_cases','m_avg_transaction_amount','m_min_transaction_amount','m_max_transaction_amount','m_stddev_transaction_amount','m_number_transactions','m_unique_accounts','m_prop_discover_transactions','m_prop_amex_transactions','m_prop_diners_transactions','m_prop_mastercard_transactions','m_prop_visa_transactions','a_avg_transaction_amount','a_min_transaction_amount','a_max_transaction_amount','a_stddev_transaction_amount','a_number_transactions','m_transaction_delta','a_transaction_delta']
idColumns = ['account_id', 'location_id', 'transaction_id', 'fraud_flag']


# encode categorical features
query = """
    DROP TABLE IF EXISTS madlibflowdemo.all_features_onehot, madlibflowdemo.all_features_onehot_dictionary;
    SELECT madlib.encode_categorical_variables (
        'madlibflowdemo.all_features',
        'madlibflowdemo.all_features_onehot',
        '{}',
        NULL,
        '{}',
        NULL,
        'merchant_name=TravelCenters Of America, card_type=Diners',
        NULL,
        NULL,
        TRUE
    );
""".format(','.join(categoricalFeatures),','.join(continuosFeatues + idColumns))
cur.execute(query)

query = """
    SELECT *
    FROM madlibflowdemo.all_features_onehot
    LIMIT 5
"""
cur.execute(query)

colnames = [desc[0] for desc in cur.description]

pivotFeatures = [c for c in colnames if c not in categoricalFeatures + continuosFeatues + idColumns] 

print(pivotFeatures)

pd.DataFrame(cur.fetchall(), columns=colnames)

```

<a id="fe_cats_dep"></a>

```python
def bar_plot_groupby(data, title,x,xLabel,y,yLabel,groupby,color=None,axisRotation=90):

    # Bar plot
    pylab.rcParams['figure.figsize'] = (12, 8)
    seq_col_brew = sns.color_palette("Blues_r", 1)
    sns.color_palette(seq_col_brew)
    if color != None:
        plt = sns.barplot(x=x, y=y, data=data, color=color, hue=groupby)
    else:
        plt = sns.barplot(x=x, y=y, data=data, hue=groupby)
        
    # titles
    plt.set_title(title,fontsize=30)
    plt.set_xlabel(xLabel,fontsize=16)
    plt.set_ylabel(yLabel,fontsize=16)
    
    # rotate x axis labels
    for item in plt.get_xticklabels():
        item.set_rotation(axisRotation)

    # remove scientific notation
    plt.ticklabel_format(style='plain', axis='y')

def get_cat_gb_data_frame(col):
    query = """
        SELECT *
              ,round((record_count * 100.0) / sum(record_count) OVER(PARTITION BY col),2) AS perc_records
        FROM (
            SELECT "{}" AS col
                  ,fraud_flag
                  ,count(*) AS record_count
            FROM madlibflowdemo.all_features_onehot
            GROUP BY 1,2
        ) foo
        ORDER BY 1,2
    """.format(col)
    cur.execute(query)

    colnames = [desc[0] for desc in cur.description]
    return pd.DataFrame(cur.fetchall(), columns=colnames)
    
def on_cat_gb_selection(res):
    if res['type'] == 'change' and res['name'] == 'value':
        ipd.clear_output()
        printmd("-----\n **Select Column:**")
        ipd.display(catGPDropdown)
        df = get_cat_gb_data_frame(res['new'])
        bar_plot_groupby(df,res['new'],"col",res['new'],"perc_records","% Class Records", "fraud_flag")
    
catGPDropdown = widgets.Dropdown(
    options=pivotFeatures,
    value=pivotFeatures[0],
    description='Column:',
    disabled=False,
)

catGPDropdown.observe(on_cat_gb_selection)
printmd("-----\n **Select Column:**")
ipd.display(catGPDropdown)
df = get_cat_gb_data_frame(pivotFeatures[0])
bar_plot_groupby(df,pivotFeatures[0],"col",pivotFeatures[0],"perc_records","% Class Records","fraud_flag")
```

<a id="fe_chi_sq"></a>
#### Chi-squared testing

https://en.wikipedia.org/wiki/Chi-squared_test

https://madlib.apache.org/docs/latest/group__grp__stats__tests.html

```python
def chi2_gof_test(feature_name, response):
    query = """
        WITH freq AS (
            SELECT "{feature_name}" AS f
                  ,{response}
                  ,count(*) AS observed
            FROM madlibflowdemo.all_features_onehot
            GROUP BY 1,2
        )
        SELECT $q${feature_name}$q$ AS feature_name
              ,'{response}' AS response
              ,(madlib.chi2_gof_test(observed, expected, deg_freedom)).*
        FROM (
            SELECT observed
                  ,sum(observed) OVER (PARTITION BY f)::DOUBLE PRECISION
                       * sum(observed) OVER (PARTITION BY {response}) AS expected
            FROM freq
        ) l, (
            SELECT (count(distinct f) - 1) * (count(distinct {response}) - 1) AS deg_freedom
            FROM freq
        ) r;
    """.format(feature_name=feature_name, response=response)
    cur.execute(query)

    colnames = [desc[0] for desc in cur.description]
    return pd.DataFrame(cur.fetchall(), columns=colnames)

def chi2_gof_test_multi(feature_list, response):
    res = chi2_gof_test(feature_list[0], response)
    for i in range(1,len(feature_list)):
        res = res.append(chi2_gof_test(feature_list[i], response))
        
    return res

chi2_results = chi2_gof_test_multi(pivotFeatures, 'fraud_flag')
chi2_results.sort_values('phi', inplace=True)
ipd.display(chi2_results)
bar_plot(chi2_results, "Chi-Squared Testing","feature_name","Feature Name","phi","Test Statistic", "#4378E2")
```

```python
def sample_scatter(title, x, xLabel, y, yLabel, sampleSize):
    
    pylab.rcParams['figure.figsize'] = (8, 8)

    # Grab sample
    query = """
        SELECT count(*) AS n
        FROM madlibflowdemo.all_features_onehot;
    """
    cur.execute(query)

    colnames = [desc[0] for desc in cur.description]
    n = pd.DataFrame(cur.fetchall(), columns=colnames)['n'][0]
    limit = math.floor(n * sampleSize)
    
    query = """
        SELECT {} AS col1
              ,{} AS col2
        FROM madlibflowdemo.all_features_onehot
        LIMIT {};
    """.format(x, y, limit)
    cur.execute(query)

    colnames = [desc[0] for desc in cur.description]
    sample = pd.DataFrame(cur.fetchall(), columns=colnames)    
    
    # Generate scatterplot
    if x == y:
        sample
    plt = sns.regplot(x="col1", y="col2", data=sample)
    
    # titles
    plt.set_title("\n".join(tw.wrap(title,50)),fontsize=16)
    plt.set_xlabel(xLabel,fontsize=16)
    plt.set_ylabel(yLabel,fontsize=16)

    # add 1000s commas
    plt.get_yaxis().set_major_formatter(
        matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))
    plt.get_xaxis().set_major_formatter(
        matplotlib.ticker.FuncFormatter(lambda y, p: format(int(y), ',')))
```

```python
x = continuosFeatues[0]
y = continuosFeatues[1]

def reset():
    ipd.clear_output()
    printmd("-----\n **Select Features:**")
    ipd.display(scatterDropdown1, scatterDropdown2)

    sample_scatter("{} by {}".format(x, y), x, x, y, y, 0.05)   
    
def os1(res):
    global x
    if res['type'] == 'change' and res['name'] == 'value':
        continuosFeatues.append(x)
        x = res['new']
        continuosFeatues.remove(x)
        reset()

def os2(res):
    global y
    if res['type'] == 'change' and res['name'] == 'value':
        y = res['new']
        reset()
                   
scatterDropdown1 = widgets.Dropdown(
    options=continuosFeatues,
    value=x,
    description='x:',
    disabled=False,
)

scatterDropdown2 = widgets.Dropdown(
    options=continuosFeatues,
    value=y,
    description='y:',
    disabled=False,
)

scatterDropdown1.observe(os1)
scatterDropdown2.observe(os2)

reset()
```

## MADlib Flow - Model Inference


### Cache table functions

```python
%%sql

-- Create merchants table for caching
DROP FUNCTION IF EXISTS madlibflowdemo.mf_merchants(text);
CREATE OR REPLACE FUNCTION madlibflowdemo.mf_merchants(out_table text)
RETURNS void AS
$body$

DECLARE
    query TEXT;
BEGIN
    query := $$
                DROP TABLE IF EXISTS $$|| out_table ||$$;
                CREATE TABLE $$|| out_table ||$$ AS
                SELECT location_id
                      ,count(CASE WHEN fraud_flag = true THEN 1 ELSE null END) AS m_fraud_cases
                      ,avg(transaction_amount) AS m_avg_transaction_amount
                FROM madlibflowdemo.locations
                LEFT JOIN madlibflowdemo.transactions
                USING (location_id)
                WHERE transaction_date > now() - interval '30 days'
                GROUP BY location_id;
            $$;

    EXECUTE query;
END;
    
$body$ 
LANGUAGE 'plpgsql';

-- test function
SELECT madlibflowdemo.mf_merchants('madlibflowdemo.mf_merchants');

SELECT count(*) FROM madlibflowdemo.mf_merchants;
```

```python
%%sql

-- Create accounts table for caching
DROP FUNCTION IF EXISTS madlibflowdemo.mf_accounts(text);
CREATE OR REPLACE FUNCTION madlibflowdemo.mf_accounts(out_table text)
RETURNS void AS
$body$
DECLARE
    query TEXT;
BEGIN
    query := $$
                DROP TABLE IF EXISTS $$|| out_table ||$$;
                CREATE TABLE $$|| out_table ||$$ AS
                SELECT account_id
                      ,avg(transaction_amount) AS a_avg_transaction_amount
                FROM madlibflowdemo.accounts
                LEFT JOIN madlibflowdemo.transactions
                USING (account_id)
                WHERE transaction_date > now() - interval '30 days'
                GROUP BY account_id;
            $$;

    EXECUTE query;
END;            
$body$ 
LANGUAGE 'plpgsql';

-- test function
SELECT madlibflowdemo.mf_accounts('madlibflowdemo.mf_accounts');

SELECT count(*) FROM madlibflowdemo.mf_accounts;
```

### Feature Engine Code


```python
%%sql

DROP TABLE IF EXISTS madlibflowdemo.feature_engine_test;
CREATE TABLE madlibflowdemo.feature_engine_test AS

SELECT t.*
      ,m.m_fraud_cases
      ,abs(t.transaction_amount - m.m_avg_transaction_amount) AS m_transaction_delta
      ,abs(t.transaction_amount - a.a_avg_transaction_amount) AS a_transaction_delta
      ,CASE WHEN t.merchant_state = 'RS' THEN 1 ELSE 0 END AS merchant_state_2
FROM madlibflowdemo.transactions AS t
JOIN madlibflowdemo.mf_merchants AS m
USING (location_id)
JOIN madlibflowdemo.mf_accounts AS a
USING (account_id);
```

### Predict function


```python
%%sql

-- Model scoring
DROP TABLE IF EXISTS madlibflowdemo.score_test;
SELECT madlib.forest_predict('madlibflowdemo.rf_model',
                             'madlibflowdemo.feature_engine_test',
                             'madlibflowdemo.score_test',
                             'prob');
                
-- Join score data back with original
DROP TABLE IF EXISTS madlibflowdemo.score_test_tmp;
CREATE TABLE madlibflowdemo.score_test_tmp AS
SELECT *
FROM madlibflowdemo.score_test
JOIN madlibflowdemo.feature_engine_test
USING (transaction_id);
DROP TABLE madlibflowdemo.score_test;
ALTER TABLE madlibflowdemo.score_test_tmp RENAME TO score_test;
```
